```python
import requests
```

```python

def get_poi_info(poi_name, api_key):
    # 高德地图POI搜索API的基础URL
    url = "https://restapi.amap.com/v3/place/text"

    # 请求参数
    params = {
        'keywords': poi_name,
        'key': api_key,
        'city': '拉萨',  # 指定城市为拉萨
        'output': 'JSON'
    }

    # 发起请求
    response = requests.get(url, params=params)
    # print(response.json())
    
    # 解析响应
    if response.status_code == 200:
        data = response.json()
        if data['status'] == '1' and int(data['count']) > 0:
            pois = data['pois']
            for poi in pois:
                name = poi.get('name', 'N/A')
                address = poi.get('address', 'N/A')
                telephone = poi.get('tel', 'N/A')
                print(f"Name: {name}\nAddress: {address}\nTelephone: {telephone}\n")
        else:
            print(f"No results found for {poi_name}.")
    else:
        print(f"Request failed with status code {response.status_code}.")

# 示例调用
api_key = "d099c94cd9b0b54238c89b7294a96ef9"
poi_name = "西藏青年国际旅行社有限公司"
get_poi_info(poi_name, api_key)
```

```python

def get_poi_info_in_lhasa(api_key):
    # 高德地图POI搜索API的基础URL
    url = "https://restapi.amap.com/v3/place/text"

    # 请求参数
    params = {
        'keywords': '旅行社',
        'city': '拉萨',
        'key': api_key,
        'output': 'JSON',
        'citylimit': 'true'  # 限制在指定城市内搜索
    }

    # 发起请求
    response = requests.get(url, params=params)
    
    # 解析响应
    if response.status_code == 200:
        data = response.json()
        if data['status'] == '1' and int(data['count']) > 0:
            pois = data['pois']
            for poi in pois:
                name = poi.get('name', 'N/A')
                address = poi.get('address', 'N/A')
                telephone = poi.get('tel', 'N/A')
                # 只打印有电话的旅行社
                if telephone != 'N/A':
                    print(f"Name: {name}\nAddress: {address}\nTelephone: {telephone}\n")
        else:
            print("No results found.")
    else:
        print(f"Request failed with status code {response.status_code}.")

# 示例调用
api_key = "d099c94cd9b0b54238c89b7294a96ef9"
get_poi_info_in_lhasa(api_key)
```

```python
import requests
import time

def get_poi_info_in_lhasa(api_key, query):
    url = "https://restapi.amap.com/v3/place/text"
    
    # 拉萨市的经纬度范围，可以调整精度以减少网格点
    min_lng, max_lng = 91.0, 92.0
    min_lat, max_lat = 29.5, 30.5
    grid_size = 0.1  # 网格大小

    pois = []
    
    for lng in range(int((max_lng - min_lng) / grid_size)):
        for lat in range(int((max_lat - min_lat) / grid_size)):
            # 网格中心点
            center_lng = min_lng + lng * grid_size + grid_size / 2
            center_lat = min_lat + lat * grid_size + grid_size / 2
            
            # 请求参数
            params = {
                'keywords': query,
                'location': f'{center_lng},{center_lat}',
                'key': api_key,
                'output': 'JSON',
                'city': '拉萨',
                'radius': int(grid_size * 100000),  # 将网格大小转换为米
                'offset': 20  # 每次获取20条数据
            }
            
            response = requests.get(url, params=params)
            if response.status_code == 200:
                data = response.json()
                if data['status'] == '1' and int(data['count']) > 0:
                    pois.extend(data['pois'])
                time.sleep(0.5)  # 添加延迟避免频繁请求
            else:
                print(f"Request failed with status code {response.status_code} for location {center_lng},{center_lat}")
    
    # 去除重复项
    unique_pois = {poi['id']: poi for poi in pois}.values()
    return unique_pois

def filter_pois_with_telephone(pois):
    filtered_pois = []
    for poi in pois:
        name = poi.get('name', 'N/A')
        address = poi.get('address', 'N/A')
        telephone = poi.get('tel', 'N/A')
        if telephone != 'N/A'and telephone != '[]':
            filtered_pois.append({'name': name, 'address': address, 'telephone': telephone})
    return filtered_pois

def main():
    api_key = "d099c94cd9b0b54238c89b7294a96ef9"
    query = "旅行社"
    pois = get_poi_info_in_lhasa(api_key, query)
    filtered_pois = filter_pois_with_telephone(pois)
    for poi in filtered_pois:
        print(f"Name: {poi['name']}\nAddress: {poi['address']}\nTelephone: {poi['telephone']}\n")

if __name__ == "__main__":
    main()
```
