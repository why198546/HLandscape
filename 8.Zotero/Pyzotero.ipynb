```python
import requests
from pyzotero import zotero

# 初始化 Pyzotero 对象
library_id = '9769982'
api_key = 'fw2EPYr4Dby6PulXbrftoSZ2'
zot = zotero.Zotero(library_id, "user", api_key)

# 使用 CrossRef API 获取元数据
doi = "10.1016/j.isprsjprs.2023.04.005"
crossref_url = f"https://api.crossref.org/works/{doi}"
response = requests.get(crossref_url)
metadata = response.json()['message']

# 创建新项的模板
new_item_template = zot.item_template('journalArticle')  # 以期刊文章为例
new_item_template['title'] = metadata.get('title')[0]
new_item_template['DOI'] = doi
new_item_template['creators'] = [{"creatorType": "author", "name": author['family']} for author in metadata.get('author', [])]

# 提交新项
new_item = zot.create_items([new_item_template])

# 检查是否成功添加
if new_item['successful']:
    print("Item successfully added with DOI:", doi)
else:
    print("Failed to add item.")
```

```python
import requests
from pyzotero import zotero

def add_items_to_zotero_by_doi(doi_list, library_id, api_key, collection_id):
    # 初始化 Pyzotero 对象
    zot = zotero.Zotero(library_id, "user", api_key)
    
    for doi_url in doi_list:
        # 去除 "https" 前缀
        doi = doi_url.replace("https://doi.org/", "")
        
        # 使用 CrossRef API 获取元数据
        crossref_url = f"https://api.crossref.org/works/{doi}"
        response = requests.get(crossref_url)
        metadata = response.json()['message']
        
        # 创建新项的模板
        new_item_template = zot.item_template('journalArticle')  # 以期刊文章为例
        new_item_template['title'] = metadata.get('title')[0]
        new_item_template['DOI'] = doi
        new_item_template['creators'] = [{"creatorType": "author", "name": author['family']} for author in metadata.get('author', [])]
        
        # 提交新项
        new_item = zot.create_items([new_item_template])
        
        # 将新项添加到特定集合
        if new_item['successful']:
            zot.addto_collection(collection_id, new_item['successful'].values())
            print(f"Item successfully added to collection with DOI: {doi}")
        else:
            print(f"Failed to add item with DOI: {doi}")
```

```python

# 使用示例
doi_list = ["https://doi.org/10.1234/example.doi1", "https://doi.org/10.1234/example.doi2"]
library_id = "Your_Library_ID"
api_key = "Your_API_Key"
collection_id = "Your_Collection_ID"
add_items_to_zotero_by_doi(doi_list, library_id, api_key, collection_id)
```

```python
def add_item_by_doi(doi):
    try:
        # Retrieve the metadata for the item from Zotero by DOI
        # This will return a list of one or more dict items
        items = zot.create_items([{"DOI": doi}])
        
        # Check if the items have been added successfully
        if items:
            print(f'Successfully added item with DOI: {doi}')
            return items
        else:
            print(f'No items added for DOI: {doi}')
            return None
    except Exception as e:
        print(f'An error occurred: {e}')
        return None
```
