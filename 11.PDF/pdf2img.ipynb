```python
import os
from pdf2image import convert_from_path
from PIL import Image

def convert_pdf_to_jpg(folder_path, output_folder, poppler_path):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for file_name in os.listdir(folder_path):
        if file_name.endswith('.pdf'):
            pdf_path = os.path.join(folder_path, file_name)
            images = convert_from_path(pdf_path, dpi=300, poppler_path=poppler_path)

            for i, image in enumerate(images):
                jpg_file_name = f"{os.path.splitext(file_name)[0]}_page_{i + 1}.jpg"
                jpg_path = os.path.join(output_folder, jpg_file_name)
                image.save(jpg_path, 'JPEG', quality=100)
                print(f"Converted {pdf_path} to {jpg_path}")

if __name__ == "__main__":
    folder_path = 'D:\\Documents\\审图\\新建文件夹'  # Change this to your PDF folder path
    output_folder = 'D:\\Documents\\审图\\输出文件夹'  # Change this to your desired output folder
    poppler_path = r'D:\Downloads\poppler-24.02.0\Library\bin'  # Change this to the path where Poppler is installed
    convert_pdf_to_jpg(folder_path, output_folder, poppler_path)
```
