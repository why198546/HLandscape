```python
import pandas as pd
import geopandas as gpd
import fiona
from shapely.geometry import Point
from pyautocad import Autocad,APoint
```

```python
# Set autocad environment and set an empty list to store material name

acad = Autocad(create_if_not_exists=True)
doc = acad.ActiveDocument
```

# 获取小品标注，并生成df

```python
pt1 = APoint(0,0,0)
pt2 = APoint(10000,0,0)
```

```python
figure = []
i= 0
for entity in acad.iter_objects(['TDbSymbIndexPointer'],dont_cast=True):
    if '小品' in entity.UpText or '亭' in entity.UpText or '廊架' in entity.UpText:
        i+=1
        figure.append([entity.UpText,entity.DownText,entity.IndexLabel,entity.DrawingLabel,i]) #上标文字，下标文字，索引编号，图号
        entity.color = i
        entity.move(pt1,pt2)
```

```python
figure
```

```python
df = pd.DataFrame(figure)
df
```

```python
df.columns = ['figure','note','label','number','Color']
df
```

```python
count = df['figure'].str.contains('钢板').sum()
count
```

```python
import os
from pdf2image import convert_from_path

# 指定包含PDF文件的文件夹
pdf_dir = r"D:\Documents\WeChat Files\wxid_3oemgw81j4rc22\FileStorage\File\2023-07\堆龙滨河景观项目-初步设计图纸部分-PDF20230417发李景-0425发兰总-0515发兰总\堆龙滨河景观项目-初步设计图纸部分-PDF20230417发李景-0425发兰总-0515发兰总"

# 遍历文件夹中的所有PDF文件
for filename in os.listdir(pdf_dir):
    if filename.startswith('XS') and filename.endswith(".pdf"):
        # 获取PDF文件的完整路径
        pdf_path = os.path.join(pdf_dir, filename)
        
        # 将PDF转换为图片
        images = convert_from_path(pdf_path)
        
        # 遍历所有生成的图片
        for image in images:
            # 为每个图片指定一个文件名（基于原始PDF文件的文件名）
            image_filename = f"{os.path.splitext(filename)[0]}.jpg"
            result = image_filename.split("-", 2)[0:2]
            image_filename = "-".join(result)+".jpg"
            # 将图片保存为JPG文件
            image.save(os.path.join(pdf_dir, image_filename), "JPEG")
```
