```python
from pyautocad import Autocad, APoint
import math
```

```python
# 初始化AutoCAD
acad = Autocad()

# 定义参数和计算半径
length = 20
R = math.sqrt(3) * length / (math.sqrt(3) * math.cos(math.pi / 12) - math.sin(math.pi / 12))
r = 2 * math.sin(math.pi / 12) * R / math.sqrt(3)*0.8734

# 创建中心圆
center = APoint(0, 0)
acad.model.AddCircle(center, R)
acad.model.AddCircle(center, R*0.9)

# 创建12个圆弧
for i in range(12):
    angles = [(i * 30-15, (i + 1) * 30-15)]
    for angle_start, angle_end in angles:
        angle_start_rad = math.radians(angle_start)
        angle_end_rad = math.radians(angle_end)
        
        # 计算圆弧的中心点
        arc_center_x = R * math.cos((angle_start_rad + angle_end_rad) / 2)
        arc_center_y = R * math.sin((angle_start_rad + angle_end_rad) / 2)
        arc_center = APoint(arc_center_x, arc_center_y)
        
        # 添加圆弧
        acad.model.AddArc(arc_center, r, (261.406 + 30*i)*pi/180, (98.594 + 30*i)*pi/180)
        acad.model.AddArc(arc_center, r*0.9, (261.406 + 30*i)*pi/180, (98.594 + 30*i)*pi/180)

# 添加文本
text_point = APoint(-7.75, -2.5)  # 调整坐标以使文本居中
text = acad.model.AddText('中秋快乐', text_point, 5)
text.Alignment = 4
```
